```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
from colorthief import ColorThief
```

```python
SIZE = 224
path = 'dataset_v0/91jUzDDrjwL._AC_SX679_.jpg'
assert(os.path.exists(path))
img = cv2.imread(path)
red_img = img.copy()

red_img = img[55:500, 215:450, :]
cv2.imwrite('temp.jpg', red_img)
plt.imshow(cv2.resize(cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB), (SIZE, SIZE)))
```

```python
# path = 'dataset_v0/91jUzDDrjwL._AC_SX679_.jpg'
path = 'temp.jpg'
color_thief = ColorThief(path)
# build a color palette
palette = color_thief.get_palette(color_count=6, quality=10)
```

```python
palette
```

```python
N = len(palette)
cols = min(8, N)
rows = (N // cols) + 1
fig = plt.figure(figsize=(14, 7)) 
for i, rgb in enumerate(palette):
    fig.add_subplot(rows, cols, i+1) 
    rgb = palette[i]
    square = np.full((5, 5, 3), rgb, dtype=np.uint8)
    plt.imshow(square)
    plt.axis('off')
```
